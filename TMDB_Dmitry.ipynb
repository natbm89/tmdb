```python
# Recolección de datos de cambios recientes en películas

# Obtener la fecha actual en formato YYYY-MM-DD
current_date = datetime.now().strftime("%Y-%m-%d")

total_pages = None

all_data = []

i = 1
while True:
    response = requests.get(CHANGES_URL + f"&page={i}")
    if response.status_code == 200:
        data = response.json()
        all_data.extend(data["results"])

        if total_pages is None:
            total_pages = data["total_pages"]
        
    else:
        print(f"Error en la solicitud: {response.status_code}")

    if i > total_pages:
        break

    i += 1
    time.sleep(1)

file_name = f"movie_changes_{current_date}.json"
with open(file_name, 'w', encoding='utf-8') as file:
    json.dump(all_data, file, ensure_ascii=False, indent=4)
```

```python
top_id = 1_518_400

new_movies = [change for change in all_data if change["id"] > top_id]
```

```python
MOVIE_DETAILS_ENDPOINT = "https://api.themoviedb.org/3/movie/{}?api_key={}&language=en-US"

new_movies_details = []
for movie in new_movies:
    response = requests.get(MOVIE_DETAILS_ENDPOINT.format(movie['id'], TMDB_KEY))
    new_movies_details.append(response.json())
    time.sleep(1)
```

```python
# ID de la película (ejemplo: 550 para "Fight Club")
movie_id = 550

# URL del endpoint de créditos
reparto_url = f"https://api.themoviedb.org/3/movie/{movie_id}/credits?api_key={TMDB_KEY}"

# Realizar la solicitud GET
response = requests.get(reparto_url)

# Verificar si la solicitud fue exitosa
if response.status_code == 200:
    data = response.json()
    
    # Extraer el reparto (cast)
    cast = data["cast"]
    
    # Guardar los datos en un archivo JSON
    file_name = f"movie_{movie_id}_credits.json"
    with open(file_name, 'w', encoding='utf-8') as file:
        json.dump(cast, file, ensure_ascii=False, indent=4)
    
    print(f"Créditos de la película guardados en '{file_name}'")
else:
    print(f"Error en la solicitud: {response.status_code}")
```
